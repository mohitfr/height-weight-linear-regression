# Ridge and Lasso Regression - Algerian Forest Fire
**Goal:** Predicting the **Fire Weather Index (FWI)** from weather and fire danger rating 

**Dataset:**  algerian_forest_fires_cleaned.csv
**Library:** scikit-learn, pandas, numpy, matplotlib, seaborn

## 1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler
import pickle
%matplotlib inline
```

## 2. Load the dataset

```python
dataset=pd.read_csv('Algerian_forest_fires_cleaned.csv')
dataset.head()
```

```python
dataset.info()
```

```python
dataset.columns
```

Drop day, month and year(Not required in the model training)

```python
df=dataset.drop(['day','month','year'], axis=1)
```

```python
df.head()
```

```python
df['Classes']=np.where(df['Classes'].str.contains("not fire"),0,1)
```

```python
df.tail()
```

```python
df.info()
```

```python
df['Classes'].value_counts()
```

## 3. Divide the dataset(Independent and Dependent features)

```python
x=df.drop('FWI', axis=1)
y=df['FWI']
```

```python
x.head()
```

```python
y.tail()
```

## 4. Split the Dataset into training and testing sets.

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25, random_state=42)
```

```python
print(f'Train : {x_train.shape[0]} samples')
print(f'Test : {x_test.shape[0]} samples')

print(f'Percentage of training samples: {x_train.shape[0]/(x_train.shape[0] + x_test.shape[0]) * 100:.2f}%')
print(f'Percentage of testing samples: {x_test.shape[0]/(x_train.shape[0] + x_test.shape[0]) * 100:.2f}%')
```

```python
x_train.corr()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(x_train.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Features')
plt.show()
```

## 5. Feature selection

If two feature are strongly correlated, we only need one of them to train our model because of indirect relation.

```python
def correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: 
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
corr_features = correlation(x_train, 0.8)
```

```python
print(f'x_train shape: {x_train.shape}')
```

```python
x_train.drop(corr_features, axis=1, inplace=True)
print(f'x_train shape after dropping correlated features: {x_train.shape}')
x_test.drop(corr_features, axis=1, inplace=True)
print(f'x_test shape after dropping correlated features: {x_test.shape}')
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(y, bins=20, kde=True, color='steelblue', alpha=0.6)
plt.title('Distribution of FWI')
plt.xlabel('FWI')
plt.ylabel('Frequency')
plt.show()
```

## 6. Feature Scaling or Standardization

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

print('Scaling done successfully.')
print(f'Feature means (should be ~0): {x_train_scaled.mean(axis=0).round(3)}')
print(f'Feature standard deviations (should be ~1): {x_train_scaled.std(axis=0).round(3)}')
```

### Box plot for original and standardized data

```python

plt.subplots(figsize=(15, 5))

plt.subplot(1, 2, 1)
sns.boxplot(data=x_train)
plt.title('x_train Before Scaling')

plt.subplot(1, 2, 2)
sns.boxplot(data=x_train_scaled)
plt.title('x_train After Scaling')
```

## 7. Train models

### 7.1 Linear Regression

```python
linear=LinearRegression()
linear.fit(x_train_scaled, y_train)

print('Linear Regression model trained successfully.\n')
print('Model Coefficients:')

for i, coef in enumerate(linear.coef_):
    print(f'  Feature {i}: {coef:.4f}')
    
print(f'\nModel Intercept: {linear.intercept_}')
```

#### Equation of Linear Regression

```python
x_test.head()
```

```python
#equation of the
equation = f'FWI = {linear.intercept_:.4f} + '
equation += ' + '.join([f'({linear.coef_[i]:.4f} * {col})' for i, col in enumerate(x_test.columns)])
print(equation)
```

The equation of our linear regression is :
FWI = 7.1560 + (-0.0363 * Temperature) + (-0.1744 * RH) + (0.0070 * Ws) + (-0.0411 * Rain) + (-0.8232 * FFMC) + (3.6983 * DMC) + (4.8406 * ISI) + (0.4042 * Classes) + (-0.3926 * Region)

The model is now trained and ready for evaluation. Next, we will evaluate the model's performance on the test set using various metrics such as Mean Squared Error (MSE), Mean Absolute Error (MAE), and R-squared (R²).

```python
linear.predict(x_test_scaled)

mse = mean_squared_error(y_test, linear.predict(x_test_scaled))
msa = mean_absolute_error(y_test, linear.predict(x_test_scaled))
rmse = np.sqrt(mse)
r2 = r2_score(y_test, linear.predict(x_test_scaled))

print(f'Mean Squared Error (MSE)       : {mse:.4f}')
print(f'Mean Absolute Error (MAE)      : {msa:.4f}')
print(f'Root Mean Squared Error (RMSE) : {rmse:.4f}')
print(f'R-Squared (R2)                 : {r2:.4f}')
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=linear.predict(x_test_scaled), y=y_test, color='blue')
plt.title('Predicted vs Actual FWI')
plt.xlabel('Predicted FWI')
plt.ylabel('Actual FWI')
```

```python
residuals = y_test - linear.predict(x_test_scaled)
plt.figure(figsize=(8, 5))
sns.histplot(residuals, bins=20, kde=True, color='steelblue', alpha=0.6)
plt.title('Distribution of Residuals')
plt.xlabel('Residuals')
plt.show()
```

### 7.2 . Ridge Regression

```python
ridge = Ridge(alpha=1.0)
ridge.fit(x_train_scaled, y_train)

print('Ridge Regression model trained successfully.\n')
print('Model Coefficients:')

for i, coef in enumerate(ridge.coef_):
    print(f'  Feature {i}: {coef:.4f}')
    
print(f'\nModel Intercept: {ridge.intercept_}')
```

#### Equation of Ridge Regression

```python
equation = f'FWI = {ridge.intercept_:.4f} + '
equation += ' + '.join([f'({ridge.coef_[i]:.4f} * {col})' for i, col in enumerate(x_test.columns)])
print(equation)
```

Equation of Ridge regression:
FWI = 7.1560 + (-0.0269 * Temperature) + (-0.2052 * RH) + (0.0255 * Ws) + (-0.0407 * Rain) + (-0.7858 * FFMC) + (3.6829 * DMC) + (4.7449 * ISI) + (0.4342 * Classes) + (-0.3847 * Region)

Now that the model is trained, we can evaluate its performance on the test set using the same metrics as before.

```python
ridge.predict(x_test_scaled)

mse_r = mean_squared_error(y_test, ridge.predict(x_test_scaled))
msa_r = mean_absolute_error(y_test, ridge.predict(x_test_scaled))
rmse_r = np.sqrt(mse_r)
r2_r = r2_score(y_test, ridge.predict(x_test_scaled))

print(f'Mean Squared Error (MSE)       : {mse_r:.4f}')
print(f'Mean Absolute Error (MAE)      : {msa_r:.4f}')
print(f'Root Mean Squared Error (RMSE) : {rmse_r:.4f}')
print(f'R-Squared (R2)                 : {r2_r:.4f}')
```

#### Visualization

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=ridge.predict(x_test_scaled), y=y_test, color='blue')
plt.title('Predicted vs Actual FWI (Ridge)')
plt.xlabel('Predicted FWI')
plt.ylabel('Actual FWI')
plt.show()
```

```python
residuals = y_test - ridge.predict(x_test_scaled)
plt.figure(figsize=(8, 5))
sns.histplot(residuals, bins=20, kde=True, color='steelblue', alpha=0.6)
plt.title('Distribution of Residuals')
plt.xlabel('Residuals')
plt.show()
```

### 7.3 . Lasso regression

```python
lasso = Lasso(alpha=1.0)
lasso.fit(x_train_scaled, y_train)

print('Lasso Regression model trained successfully.\n')
print('Model Coefficients:')

for i, coef in enumerate(lasso.coef_):
    print(f'  Feature {i}: {coef:.4f}')
    
print(f'\nModel Intercept: {lasso.intercept_}')
```

Equation of Lasso regression

```python
equation = f'FWI = {lasso.intercept_:.4f} + '
equation += ' + '.join([f'({lasso.coef_[i]:.4f} * {col})' for i, col in enumerate(x_test.columns)])
print(equation)
```

Equation of Lasso regression : FWI = 7.1560 + (0.0000 * Temperature) + (-0.0000 * RH) + (0.0000 * Ws) + (-0.0000 * Rain) + (0.0000 * FFMC) + (2.9903 * DMC) + (4.0082 * ISI) + (0.0000 * Classes) + (-0.0000 * Region)

The model is trained, we can now evaluate it as well

```python
mse_l = mean_squared_error(y_test, lasso.predict(x_test_scaled))
msa_l = mean_absolute_error(y_test, lasso.predict(x_test_scaled))
rmse_l = np.sqrt(mse_l)
r2_l = r2_score(y_test, lasso.predict(x_test_scaled))

print(f'Mean Squared Error (MSE)       : {mse_l:.4f}')
print(f'Mean Absolute Error (MAE)      : {msa_l:.4f}')
print(f'Root Mean Squared Error (RMSE) : {rmse_l:.4f}')
print(f'R-squared (R2)                 : {r2_l:.4f}')
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(y_test, lasso.predict(x_test_scaled), color='blue')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Lasso Regression: Actual vs Predicted')
plt.show()
```

```python
residuals = y_test - lasso.predict(x_test_scaled)
plt.figure(figsize=(8, 5))
sns.histplot(residuals, bins=20, kde=True, color='steelblue', alpha=0.6)
plt.title('Distribution of Residuals')
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.show()
```
